```python
# Apply label encoding on required columns
# Import label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data['Suburb'].head()
```

```python
# Use on Suburb column
# Fit
le.fit(data['Suburb'])
```

```python
# Fit transform
data['Suburb'] = le.fit_transform(data['Suburb'])
```

```python
# Check Data
data['Suburb'].head()
```

```python
# Drop Address column
data.drop(columns= 'Address', inplace=True)
data.head()
```

```python
# Apply label encoding on all remaining columns
cols = ['Type','Method','SellerG','CouncilArea','Regionname','Date']
def encode(c):
    le.fit(data[c])
    data[c]=le.fit_transform(data[c])
for c in cols:
    encode(c)
```

```python
# Check data
data.head()
```

```python
# Drop price column
X = data.drop(columns='Price')
X.head()
```

```python
X.shape
```

```python
# Split data in x and y (independent features and dependent features)
y = data['Price'].copy()
y.head()
```

```python
y.shape
```

# Perform train-test spliting
x - 2parts , y - 2parts

xtrain, xtest  ytrain, ytest

```python
Xtrain,Xtest = X[0:10000],X[10000:]
ytrain,ytest = y[0:10000],y[10000:]
```

```python
#Apply model-Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
dir(lr_model)
```

```python
lr_model.fit(Xtrain, ytrain)
```

```python
lr_model.intercept_
```

```python
lr_model.coef_
```

```python
# Lets predict xtest
pred = lr_model.predict(Xtest)
```

```python
pred_df = pd.DataFrame(pred)
```

```python
# Predicted value
pred_df.head()
```

```python
# Actual values
ytest.head()
```

```python
# Get error
from sklearn.metrics import mean_absolute_error,mean_squared_error

mse = mean_squared_error(ytest, pred_df)
print('Error :',mse)

mae = mean_absolute_error(ytest, pred_df)
print('Error :',mae)
```
